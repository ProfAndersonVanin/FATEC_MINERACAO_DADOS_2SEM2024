# **Explorando dados tabulares com Pandas**

NumPy fornece muitas funcionalidades e ferramentas necessárias para trabalhar com números, como matrizes de valores numéricos. Porém, quando você começa a lidar com tabelas de dados bidimensionais, o pacote Pandas oferece uma estrutura mais conveniente para trabalhar: o DataFrame.

Execute a célula a seguir para importar a biblioteca Pandas e criar um DataFrame com três colunas. A primeira coluna é uma lista de nomes de alunos e a segunda e terceira colunas são as matrizes NumPy contendo o tempo de estudo e os dados das notas.

```python
df_estudantes = pd.DataFrame({'Nome': ['Dan', 'Joann', 'Pedro', 'Rosie', 'Ethan', 'Vicky', 'Frederic', 'Jimmie',
                                     'Rhonda', 'Giovanni', 'Francesca', 'Rajab', 'Naiyana', 'Kian', 'Jenny',
                                     'Jakeem','Helena','Ismat','Anila','Skye','Daniel','Aisha'],
                            'HorasEstudo':estudantes_data[0],
                            'Notas':estudantes_data[1]})

df_estudantes
```

Observe que além das colunas especificadas, o DataFrame inclui um índice para identificar exclusivamente cada linha. Poderíamos ter especificado o índice explicitamente e atribuído qualquer tipo de valor apropriado (por exemplo, um endereço de e-mail). No entanto, como não especificamos um índice, foi criado um com um valor inteiro exclusivo para cada linha.

## **Encontrar e filtrar dados em um DataFrame**

Você pode usar o método loc do DataFrame para recuperar dados para um valor de índice específico, como este.

```python
# Pega os dados por índice com valor 5
df_estudantes.loc[5]
```

Você também pode obter os dados em vários valores de índice, como este:

```python
# Pega as linhas com índice de 0 a 5
df_estudantes.loc[0:5]
```

Além de poder usar o método loc para encontrar linhas com base no índice, você pode usar o método iloc para encontrar linhas com base em sua posição ordinal no DataFrame (independentemente do índice):

```python
# Pega dados das primeiras 5 linhas
df_estudantes.iloc[0:5]
```

Observe atentamente os resultados do iloc[0:5] e compare-os com os resultados do loc[0:5] obtidos anteriormente. Você pode ver a diferença?

O método loc retornou linhas com rótulo de índice na lista de valores de 0 a 5, que inclui 0, 1, 2, 3, 4 e 5 (seis linhas). No entanto, o método iloc retorna as linhas nas posições incluídas no intervalo de 0 a 5. Como os intervalos de números inteiros não incluem o valor do limite superior, isso inclui as posições 0, 1, 2, 3 e 4 (cinco linhas).

O iloc identifica valores de dados em um DataFrame por posição, que se estende além das linhas até as colunas. Então, por exemplo, você pode usá-lo para encontrar os valores das colunas nas posições 1 e 2 da linha 0, assim:

```python
df_estudantes.iloc[0,[1,2]]
```

Vamos voltar ao método loc e ver como ele funciona com colunas. Lembre-se de que você usa loc para localizar itens de dados com base em valores de índice em vez de posições. Na ausência de uma coluna de índice explícita, as linhas no nosso DataFrame são indexadas como valores inteiros, mas as colunas são identificadas pelo nome:

```python
df_estudantes.loc[0,'Notas']
```

Aqui está outro truque útil. Você pode usar o método loc para localizar linhas indexadas com base em uma expressão de filtragem que faz referência a colunas nomeadas diferentes do índice, como esta:

```python
df_estudantes.loc[df_estudantes['Nome']=='Aisha']
```

Na verdade, você não precisa usar explicitamente o método loc para fazer isso. Você pode simplesmente aplicar uma expressão de filtragem DataFrame, como esta:

```python
df_estudantes[df_estudantes['Nome']=='Aisha']
```

E para garantir, você pode obter os mesmos resultados usando o método query do DataFrame, assim:

```python
df_estudantes.query('Nome=="Aisha"')
```

Os três exemplos anteriores sublinham uma verdade confusa sobre como trabalhar com Pandas. Freqüentemente, existem várias maneiras de obter os mesmos resultados. Outro exemplo disso é a maneira como você se refere ao nome de uma coluna DataFrame. Você pode especificar o nome da coluna como um valor de índice nomeado (como nos exemplos df_students['Name'] que vimos até agora), ou você pode usar a coluna como uma propriedade do DataFrame, assim:

```python
df_estudantes[df_estudantes.Nome == 'Aisha']
```

## Mean e Sum
Pandas também disponibiliza várias funções estatísticas e matemáticas como `.mean()` e `.sum()` que podem ser aplicadas diretamente em um DataFrame ou Séries.

```python
valores = pd.Series([3000, 3500, 11250])
valores.mean()
```

```python
valores.sum()
```

## Columns
Usar essas funções em um DataFrame inteiro pode não fazer muito sentido, nesse caso podemos direcionar a função individualmente para uma coluna. O método .columns mostrará todas as colunas de um DataFrame:

Selecionando uma coluna com dados numéricos podemos utilizar a função estatística `.mean()` novamente:

# **Visualizando e selecionando dados**
Na prática, realizar alterações nos dados e visualizá-los constantemente é uma tarefa de rotina. Nem sempre precisamos alterar todos os dados de um DataFrame, portanto veremos maneiras diferentes de selecionar.

## Head
Para visualizar as primeiras 5 linhas do seu DataFrame existe a função `.head()` que é muito utilizada:

```python
# exibindo as primeiras 5 linhas do DataFrame
df_carros.head()
```

## Tail
Outro método muito utilizado é o `.tail()` que seleciona as últimas 5 linhas do seu DataFrame:

```python
# exibindo as últimas 5 linhas do DataFrame
df_carros.tail()
```

Ao selecionar colunas podemos elaborar filtros condicionais entre colchetes, vejamos a sintaxe:

```python
# selecionando carros com quilometragem acima de 100.000
df_carros[df_carros["Quilometragem"] > 100000]
```

```python
# selecionando apenas os carros da Honda
df_carros[df_carros["Fabricante"] == "Honda"]
```

## Groupby
Para comparar mais colunas no contexto de outra coluna, podemos usar o `.groupby()`, por exemplo, podemos agrupar os dados por fabricante e calcular a média das colunas numéricas:

```python
# agrupando os dados pelo fabricante e calculando a média das outras colunas (Quilometragem por exemplo)
df_carros.groupby("Fabricante")['Quilometragem'].mean()
```
